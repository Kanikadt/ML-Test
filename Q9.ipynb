```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.5,100.5,0.5)
```

for y=x

```python
# for y=x
y1 = x
plt.plot(x, y1)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=x")
plt.grid(True)
plt.show()
```

```python
#for y=x^2
def f_2(x):
    return x**2
y2 = f_2(x)
plt.plot(x,y2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=x^2")
plt.grid(True)
plt.show()
```

```python
#for y=(x^3)/100
def f_3(x):
    return (x**3)/100
x = np.arange(0.5,100.5,0.5)
y3 = f_3(x)
plt.plot(x,y3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=(x^3)/100")
plt.grid(True)
plt.show()
```

```python
#for y=sin(x)
def f_4(x):
    return np.sin(x)
y4 = f_4(x)
plt.plot(x,y4)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=sin(x)")
plt.grid(True)
plt.show()
```

```python
#for y=sin(x)/x
def f_5(x):
    return (np.sin(x))/x
y5 = f_5(x)
plt.plot(x,y5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=sin(x)/x")
plt.grid(True)
plt.show()
```

```python
#for y=log(x)
def f_6(x):
    return np.log(x)
x = np.arange(0.5,100.5,0.5)
y6 = f_6(x)
plt.plot(x,y6)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=log(x)")
plt.grid(True)
plt.show()
```

```python
#for y=e^x
def f_7(x):
    return np.exp(x)
x = np.arange(0.5,100.5,0.5)
y7 = f_7(x)
plt.plot(x,y7)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of y=e^x")
plt.grid(True)
plt.show()
```
